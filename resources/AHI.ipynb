```python
# Import data into a Pandas DataFrame
Affordable_Housing_Inventory_20240703_df = pd.read_csv(r'C:\Users\aspyn\affordable_housing\Affordable_Housing_Inventory_20240703.csv')

# Drop unwanted columns
columns_to_drop = ['PSH_Units', 'MFI_20', 'MFI_30', 'MFI_40', 'MFI_50', 
                   'MFI_60', 'MFI_65', 'MFI_70', 'MFI_80', 'MFI_100', 'MFI_120',
                   'FIL_Status', 'CAL_FIL', 'Date_FIL_Recorded', 'Dev_Agreement',
                   'Density_Bonus', 'FIL', 'LIHTC', 'GO_Bonds_2006', 'GO_Bonds_2013',
                   'GO_Bonds_2018', 'GO_Bonds_2022', 'CDBG', 'CIP_Funds', 'HOME',
                   'HTF', 'Project_Connect', 'City_Subsidy']
Affordable_Housing_Inventory_20240703_df.drop(columns=columns_to_drop, inplace=True)

# Calculate the total number of affordable units across all districts (Council District 1 to 10)
total_affordable_units = 0
for district in range(1, 11):
    total_units = Affordable_Housing_Inventory_20240703_df[
        Affordable_Housing_Inventory_20240703_df["Council_District"] == str(district)
    ]["Affordable_Units"].sum()
    total_affordable_units += total_units

# Calculate the percentage of affordable units
total_units = len(Affordable_Housing_Inventory_20240703_df)
percentage_affordable = (total_affordable_units / total_units) * 100

# Convert total_affordable_units to Python int
total_affordable_units = int(total_affordable_units)

# Convert DataFrame to a list of dictionaries
data_to_insert = Affordable_Housing_Inventory_20240703_df.to_dict(orient='records')

# Connect to MongoDB
client = MongoClient('localhost', 27017)
db = client['mydatabase']

# Collection for affordable housing details
collection_affordable_housing = db['affordable_housing']
result_affordable_housing = collection_affordable_housing.insert_many(data_to_insert)
print(f"Inserted {len(result_affordable_housing.inserted_ids)} documents into 'affordable_housing' collection")

# Collection for statistics or aggregated data
statistics_data = {
    'total_units': total_units,
    'total_affordable_units': total_affordable_units,
    'percentage_affordable': percentage_affordable
}

collection_statistics = db['housing_statistics']
result_statistics = collection_statistics.insert_one(statistics_data)
print(f"Inserted document into 'housing_statistics' collection")

# Close MongoDB connection
client.close()
```
